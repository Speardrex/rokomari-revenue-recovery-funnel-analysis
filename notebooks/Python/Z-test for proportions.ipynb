```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

df = pd.read_csv('ab_test.csv')

df['variant'] = df['variant'].astype(str).str.strip()

metrics = {
    'Click-Through Rate': {'success': 'clicks', 'total': 'sessions'},
    'Order Conversion Rate': {'success': 'orders', 'total': 'sessions'}
}

for name, cols in metrics.items():
    success_a = df[df['variant'] == 'A'][cols['success']].sum()
    total_a = df[df['variant'] == 'A'][cols['total']].sum()

    success_b = df[df['variant'] == 'B'][cols['success']].sum()
    total_b = df[df['variant'] == 'B'][cols['total']].sum()

    rate_a = success_a / total_a if total_a > 0 else 0
    rate_b = success_b / total_b if total_b > 0 else 0

    count = [success_b, success_a]
    nobs = [total_b, total_a]

    z_score, p_val = proportions_ztest(count, nobs, alternative='larger')

    #Final Output
    print(f"\n{name}")
    print(f"A | {success_a} | {rate_a:.2%}")
    print(f"B | {success_b} | {rate_b:.2%}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_val:.4f}")

```
